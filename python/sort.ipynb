```python
def generate_pattern(size):
    if size < 7 or size % 2 == 0:
        raise ValueError("Size must be an odd number greater than or equal to 7.")

    # Initialize the grid with dashes
    grid = [['-' for _ in range(size)] for _ in range(size)]

    mid = size // 2

    for i in range(size):
        grid[i][i] = '*'  # Main diagonal from top-left to bottom-right
        grid[i][size - 1 - i] = '*'  # Main diagonal from top-right to bottom-left

        if i % 2 == 0:
            # Calculate middle position between the stars on the current row
            middle_pos = (i + (size - 1 - i)) // 2
            grid[i][middle_pos] = '*'

            # Place stars along diagonals from the middle position
            for offset in range(1, mid + 1):
                if 0 <= i - offset < size and 0 <= middle_pos - offset < size:
                    grid[i - offset][middle_pos - offset] = '*'
                if 0 <= i - offset < size and 0 <= middle_pos + offset < size:
                    grid[i - offset][middle_pos + offset] = '*'
                if 0 <= i + offset < size and 0 <= middle_pos - offset < size:
                    grid[i + offset][middle_pos - offset] = '*'
                if 0 <= i + offset < size and 0 <= middle_pos + offset < size:
                    grid[i + offset][middle_pos + offset] = '*'

    # Print the grid
    for row in grid:
        print(''.join(row))

# Example usage
generate_pattern(7)
print()
generate_pattern(11)
```

```python
def generate_base_matrix(n):
    # Initialize the base matrix with '_'
    matrix = [['_' for _ in range(n)] for _ in range(n)]
    
    # Fill the required '*' positions
    for i in range(n):
        matrix[i][i] = '*'
        if i % 2 == 0:
            matrix[i][n-i-1] = '*'
    
    return matrix

def generate_final_matrix(n):
    base_matrix = generate_base_matrix(n)
    final_matrix = [['_' for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            if base_matrix[i][j] == '*':
                final_matrix[i][j] = '*'
                # Fill the three '_' spaces if within bounds
                if i + 3 < n:
                    final_matrix[i+3][j] = '*'
                if i - 3 >= 0:
                    final_matrix[i-3][j] = '*'
                if j + 3 < n:
                    final_matrix[i][j+3] = '*'
                if j - 3 >= 0:
                    final_matrix[i][j-3] = '*'
    
    return final_matrix

def print_matrix(matrix):
    for row in matrix:
        print("".join(row))

# Example usage:
n = 11
final_matrix = generate_final_matrix(n)
print_matrix(final_matrix)
```
